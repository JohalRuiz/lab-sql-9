```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
# engine connects SQL to python
```

```python
# Check the number of rentals for each customer for May
df_rental_may = pd.read_sql_query('''select customer_id, count(customer_id) from rentals_may
group by customer_id
order by count(customer_id) desc''', engine)
df_rental_may
```

```python
df_rental_may = df_rental_may.set_index('customer_id')
df_rental_may
```

```python
df_rental_june = pd.read_sql_query('''select customer_id, count(customer_id) from rentals_june
group by customer_id
order by count(customer_id) desc''', engine)
df_rental_june
```

```python
df_rental_june = df_rental_june.set_index('customer_id')
df_rental_june
```
